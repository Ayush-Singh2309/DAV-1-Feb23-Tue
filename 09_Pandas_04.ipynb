```python
#Given a list, create a new list which stores the number as it is if it is even and stores the double of it, if it is odd

a = [4, 7, 5, 8, 9]
# b = [4, 14, 10, 8, 18]
```

```python
a = [4, 7, 5, 8, 9]

b = []

for elem in a:
  if elem%2 == 0:
    b.append(elem)
  else:
    b.append(2*elem)

print(b)
```

```python
b = [elem if elem%2 == 0 else elem*2 for elem in a]
```

```python
print(b)
```
